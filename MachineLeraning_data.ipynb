```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'data.csv'
data1 = pd.read_csv(url)
data1.hist()
plt.show
```

```python
correlationdata = data1.corr()
print(correlationdata)
```

```python
import seaborn as sns
sns.heatmap(correlationdata, annot = True)
```

```python
sns.pairplot(data1)
```

```python
data1.plot(kind = 'scatter', x = 'purchase_after_21', y = 'total', figsize=(9,9), color = 'black')
```

```python
#M hnh bc 1
from sklearn import linear_model
modeldata = linear_model.LinearRegression()
X = pd.DataFrame(data1['purchase_after_21'])
modeldata.fit(X, y = data1['total'])
print('Intercept ca m hnh: ', modeldata.intercept_)
print('Coeffient ca m hnh: ', modeldata.coef_)
print(' chnh xc ca m hnh: ', 
      modeldata.score(X,y = data1['total']))
data1.plot(kind = 'scatter', x = 'purchase_after_21', y = 'total', figsize=(9,9), 
                 color = 'black')
y_pred = modeldata.predict(X)
plt.plot(X,y_pred, color = 'blue')
```

```python
#M hnh bc 2
from sklearn import linear_model
modeldata1 = linear_model.LinearRegression()
X = pd.DataFrame([data1['purchase_after_21'], data1['purchase_after_21']**2]).T
modeldata1.fit(X, y = data1['total'])
print('Intercept ca m hnh: ', modeldata1.intercept_)
print('Coeffient ca m hnh: ', modeldata1.coef_)
print(' chnh xc ca m hnh: ', 
      modeldata1.score(X,y = data1['total']))

import numpy as np
range = np.arange(0,1,0.01)
data1.plot(kind = 'scatter', x = 'purchase_after_21', y = 'total', figsize=(9,9),
                 color = 'black', xlim=(0,1))
predictor = pd.DataFrame([range, range**2]).T
y_pred = modeldata1.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```

```python

#M hnh bc 10
from sklearn import linear_model
modelHousing10 = linear_model.LinearRegression()
X = pd.DataFrame([data1['purchase_after_21'], data1['purchase_after_21']**2, data1['purchase_after_21']**3, data1['purchase_after_21']**4,
                  data1['purchase_after_21']**5, data1['purchase_after_21']**6,data1['purchase_after_21']**7, data1['purchase_after_21']**8,
                  data1['purchase_after_21']**9, data1['purchase_after_21']**10]).T
modelHousing10.fit(X, y = data1['total'])
print('Intercept ca m hnh: ', modelHousing10.intercept_)
print('Coeffient ca m hnh: ', modelHousing10.coef_)
print(' chnh xc ca m hnh: ', 
      modelHousing10.score(X,y = data1['total']))

import numpy as np
range = np.arange(0,1,0.1)
data1.plot(kind = 'scatter', x = 'purchase_after_21', y = 'total', figsize=(9,9),
                 color = 'black', xlim=(0,1))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5,
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = modelHousing10.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```
