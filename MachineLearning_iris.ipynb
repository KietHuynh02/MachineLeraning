```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'iris.csv'
datairis = pd.read_csv(url)
datairis.hist()
plt.show
```

```python
correlationiris = datairis.corr()
print(correlationiris)
```

```python
import seaborn as sns
sns.heatmap(correlationiris, annot = True)
```

```python
sns.pairplot(correlationiris)
```

```python
datairis.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize=(9,9), color = 'black')
```

```python
#M hnh bc 1
from sklearn import linear_model
modeliris = linear_model.LinearRegression()
X = pd.DataFrame(datairis['petal.width'])
modeliris.fit(X, y = datairis['petal.length'])
print('Intercept ca m hnh: ', modeliris.intercept_)
print('Coeffient ca m hnh: ', modeliris.coef_)
print(' chnh xc ca m hnh: ', 
      modeliris.score(X,y = datairis['petal.length']))
datairis.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize=(9,9), 
                 color = 'black')
y_pred = modeliris.predict(X)
plt.plot(X,y_pred, color = 'blue')
```

```python
#M hnh bc 2
from sklearn import linear_model
modeliris = linear_model.LinearRegression()
X = pd.DataFrame([datairis['petal.width'], datairis['petal.width']**2]).T
modeliris.fit(X, y = datairis['petal.width'])
print('Intercept ca m hnh: ', modeliris.intercept_)
print('Coeffient ca m hnh: ', modeliris.coef_)
print(' chnh xc ca m hnh: ', 
      modeliris.score(X,y = datairis['petal.width']))

import numpy as np
range = np.arange(0,2.7,0.1)
datairis.plot(kind = 'scatter', x = 'petal.width', y = 'petal.width', figsize=(9,9),
                 color = 'black', xlim=(0,2.7))
predictor = pd.DataFrame([range, range**2]).T
y_pred = modeliris.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```

```python
#M hnh bc 10
from sklearn import linear_model
modelabalone3 = linear_model.LinearRegression()
X = pd.DataFrame([datairis['petal.width'], datairis['petal.width']**2, datairis['petal.width']**3, datairis['petal.width']**4,
                  datairis['petal.width']**5, datairis['petal.width']**6,datairis['petal.width']**7, datairis['petal.width']**8,
                  datairis['petal.width']**9, datairis['petal.width']**10]).T
modelabalone3.fit(X, y = datairis['petal.length'])
print('Intercept ca m hnh: ', modelabalone3.intercept_)
print('Coeffient ca m hnh: ', modelabalone3.coef_)
print(' chnh xc ca m hnh: ', 
      modelabalone3.score(X,y = datairis['petal.length']))

import numpy as np
range = np.arange(0,2.7,0.1)
datairis.plot(kind = 'scatter', x = 'petal.width', y = 'petal.length', figsize=(9,9),
                 color = 'black', xlim=(0,2.7))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5,
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = modelabalone3.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```
